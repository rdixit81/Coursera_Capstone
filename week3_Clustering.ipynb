# Scraping Wikipedia Page

```python
#scraping data from url
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
data = pd.read_html(url, header = 0)
data
```

```python
df = pd.DataFrame(data[0])
df.head()
```

```python
#Removing non assigned values from Brough column
rows = df[ df['Borough'] == 'Not assigned'].index
df.drop(rows , inplace=True)
df.head()
df
```

```python
#Grouping Neighbourhood
Group_neighbour = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(','.join).reset_index()
Group_neighbour
```

```python
#Postcode with no neighbour
No_Neighbour = df[ df['Neighbourhood'] == 'Not assigned'].index
df['Neighbourhood'].iloc[No_Neighbour] = df['Borough'].iloc[No_Neighbour]
df.iloc[No_Neighbour]
```

```python
#Importing cordinate data
coordinates=pd.read_csv('Geospatial_Coordinates.csv')
coordinates.head(5)
```

```python
#Joining cooedinate and df dataframe
final_df=Group_neighbour.join(coordinates,lsuffix='_caller', rsuffix='_other')
```

```python
final_df
```

```python
del final_df['Postal Code']
```

```python
cluster_df = final_df[final_df['Borough'].str.contains("Toronto")].reset_index(drop=True)
cluster_df.head()
```

```python
from geopy.geocoders import Nominatim
address = 'Toronto'
geolocator = Nominatim()
location = geolocator.geocode(address, timeout=10)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
import folium
map_Toronto = folium.Map(location=[latitude, longitude], zoom_start=15)

# add markers to map
for lat, lng, label in zip(cluster_df['Latitude'], cluster_df['Longitude'], cluster_df['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='black',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.9,
        parse_html=False).add_to(map_Toronto)  
    
map_Toronto
```
